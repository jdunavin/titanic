### 3.1.1 Process titles

```python
def parse_des(x):
    
    # this line is as follows (regular expressions):
        # r signifies what follows should be treated as a raw string (no escapes necessary)
        # , ignores anything before the comma in the name; the title is right after it
        # \s matches any spaces after the comma
        # \w+. matches any characters between the spaces and the period (where the title is)
    x=re.findall(r',\s\w+.',x)
    # this line is as follows:
        # r again signifies a raw string
        # \w+ looks in the modified string above and ignores any spaces or special characters (comma, period)
    return (re.findall(r'\w+',str(x)))[0]
train2['desig']=train2['Name'].apply(parse_des)
```
